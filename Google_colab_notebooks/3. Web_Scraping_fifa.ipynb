#Inicio de Webscraping en página “fifaindex.com”


```
#Crear variables para construcción del Dataset
equipos=[]
liga=[]
ataque=[]
medio=[]
defensa=[]
promedio=[]
año=[]
```

#fifa 2005

```
for pagina in list(range(1,16)):
  url='https://www.fifaindex.com/es/teams/fifa05_1/?page='+str(pagina)
  page = requests.get(url)
  soup = BeautifulSoup(page.content, 'html.parser' )
  eq = soup.find_all('table',class_='table table-striped table-teams')
  print(url)
  for equipo in (eq[0].find_all('tbody')[0].find_all('tr')):
      selector=0
      for caracteristica in equipo.find_all('td'):
#          print(selector)
          if selector == 0:
            selector = selector+1
            continue
          elif selector == 1:
            equipos.append(caracteristica.text)
          elif selector == 2:
            liga.append(caracteristica.text)
          elif selector == 3:
            ataque.append(caracteristica.text)
          elif selector == 4:
            medio.append(caracteristica.text)
          elif selector == 5:
            defensa.append(caracteristica.text)
          elif selector == 6:
            promedio.append(caracteristica.text)
          elif selector == 7:
            año.append(2005)
          selector=selector+1
        
```

```
print(len(equipos),len(liga),len(ataque),len(medio),len(defensa),len(promedio),len(año))
#print(equipos)
```

#Creación del Dataframe 
Se crea Dataframe y se verifica que no contenga valores “NaN”.

```
df = pd.DataFrame({'equipo':equipos,'liga':liga,'ataque':ataque,'medio':medio,'defensa':defensa,'promedio':promedio,'año':año})
```

```
df["ataque"]=pd.to_numeric(df.ataque, downcast='integer')
df["medio"]=pd.to_numeric(df.ataque, downcast='integer')
df["defensa"]=pd.to_numeric(df.ataque, downcast='integer')
df["promedio"]=pd.to_numeric(df.ataque, downcast='integer')
```
